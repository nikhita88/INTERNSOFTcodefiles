```python
import math
def prime(num):
    if num % 2 == 0 and num > 2:
        return False
    for i in range(3, int(math.sqrt(num)) +1, 2):
        if num % i == 0:
            return False
    return True
print(prime(11))
```

```python
def cubes(c):
    for n in range(c):
        yield n**3
```

```python
for x in cubes(8):
    print(x)
```

```python
def fib(f):
    a = 1
    b = 1
    for i in range(f):
        yield a
        a,b = b,a+b
```

```python
for i in fib(9):
    print(i)
```

```python
def fib(f):
    a = 1
    b = 1
    output = []
   

    for i in range(f):
        output.append(a)
        a,b = b,a+b
    return output
    
```

```python
fib(9)
```
