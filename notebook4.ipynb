```python
class Student:
    def __init__(self, name, roll_no, year_of_birth):
        self.name =  name
        self.roll_no = roll_no
        self.year_of_birth = year_of_birth

    def age(self, current_year):
            return current_year - self.year_of_birth

    def __str__(self):
        return '%s %s was born in %d' % (self.name, self.roll_no, self.year_of_birth)

to = Student('nb','4', 2003)
print(to)
print(to.age(2026))
```

```python
#override
class Human(Student):
    def __init__(self, id_proof, *args, **kwargs):
        super(Human, self).__init__(*args, **kwargs)
        self.id_proof = id_proof
    def __str__(self):
        return super(Human, self).__str__() + 'and has id_proof %d' % self.id_proof

nb = Human(287366, 'nb', 'r4',2003)
print(nb)
```
