```python
try:
    file = open('marvelfile', 'r')
    file.write('iron man')
except IOError:
    print("error: couldnt find file or write data")
else:
    print("content written successfully")
```

```python
try:
    file = open('marvelfile', 'w')
    file.write('iron man')
finally:
    print('always execute finally blocks')
```

```python
def askint():
    try:
        value = int(input("enter an integer:"))
    except:
        print("uh uh!, its not an integer")
    finally:
        print("good job")
    print(value)
```

```python
def askstr():
    val = input("enter a string:")\

    if val.isalpha():
        print('done')
        print(val)
    else:
        print('not valid')


                
  
```

```python
def askint():
    while True:
        try:
            val = int(input("enter an integer:"))
        except:
            print("umm its not an integer!")
            continue
        else:
            print('thats an integer!')
            break
      
        print(val)
        
```

```python
askint()
```

```python
import sqlite3

my_db = sqlite3.connect('my db1.my_db')

#create table
my_db.execute("create table class1(id int, name text, score int)")
#insert table
my_db.execute("insert into class1(id, name, score) values(111, 'nb', 66)")
my_db.execute("insert into class1(id, name, score) values(11, 'nb', 46)")
my_db.execute("insert into class1(id, name, score) values(11, 'nb', 46)")
my_db.execute("insert into class1(id, name, score) values(131, 'cnb', 646)")
my_db.execute("insert into class1(id, name, score) values(121, 'nbv', 636)")
```

```python
my_db.commit()
```

```python
results = my_db.execute("select * from class1 order by id")
for row in results:
    print(row)
```

```python
results = my_db.execute("select * from class1 where name = 'nb' ")
for row in results:
    print(row)
```

```python
results = my_db.execute("select * from class1 where score > 111")
for row in results:
    print(row)
```

```python
results = my_db.execute("select * from class1 order by score")
for row in results:
    print(row)
```

```python
results = my_db.execute("select * from class1 order by score desc")
for row in results:
    print(row)
```
