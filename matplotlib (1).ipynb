```python
normals = pd.Series(np.random.normal(size=80))
print(normals)
normals.plot()
```

```python
normals.cumsum().plot(grid=True)
```

```python
dfvar = pd.DataFrame({'normal':np.random.normal(size = 6),
                      'gamma':np.random.gamma(1, size = 6),
                      'poisson':np.random.poisson(size=6)})
print(dfvar)
dfvar.cumsum(0).plot()
                      
```

```python
dfvar.cumsum(0).plot(subplots=True)
```

```python
dfvar.cumsum(0).plot(secondary_y='normal')
```

```python
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(10,13))
for i,var in enumerate(['normal', 'gamma', 'poisson']):
    dfvar[var].cumsum(0).plot(ax=axes[i], title=var)
axes[0].set_ylabel('cummulative sum')

    
```

```python
fig = plt.figure()
ax = plt.axes()

l = np.linspace(0,20,50)
print(l)
ax.plot(l, np.sin(l));
#plt.plot(l, np.sin(l))
```

```python
plt.plot(l, np.sin(l - 2), color = 'red')
plt.plot(l, np.sin(l - 1), color ='y')
plt.plot(l, np.sin(l - 2), color = '0.55')
```

```python
plt.plot(l, np.sin(l - 1),linestyle='-')
plt.plot(l, np.sin(l - 1), linestyle = ':')
plt.plot(l, np.sin(l - 2), linestyle='-.')
```

```python
plt.plot(l, np.sin(l - 0))
plt.xlim(-2, 8)
plt.ylim(-2,4);
```

```python
plt.plot(l, np.cos(l))
plt.title('cosine curve')
plt.xlabel('x')
plt.ylabel('.cos(x)')
```

```python
rng = np.random.RandomState(6)
x = rng.randn(80)
y = rng.randn(80)
colors = rng.rand(80)
sizes = 1000*rng.rand(80)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5,
            cmap='Blues')
plt.colorbar()
```
